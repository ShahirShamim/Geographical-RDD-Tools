```python
def nearest(pts,bds):
    bds_union = bds.geometry.unary_union
    pts['nearest'] = [nearest_points(bds_union, i)[0] for i in pts["geometry"]]
    pts['nearest_buffer'] = 0
    for i in range(0, len(pts)):
        pts["nearest_buffer"].iloc[i] = pts["nearest"].iloc[i].buffer(0.0000001)
    pts["connect"] = pts.apply(lambda row: LineString([row["geometry"], row["nearest"]]), axis=1)
    pts["distance_meters"] = pts.connect.length*100000
    return pts
```

```python
def match_to_line(pts, bds):
    matched_lines = []
    for pt_index, pt_row in pts.iterrows():
        pt_geom = pt_row['nearest_buffer']
        for line_index, line_row in bds.iterrows():
            line_geom = line_row['geometry']
            if pt_geom.intersects(line_geom):
                matched_lines.append({'pt_id': pt_index, 'line_id': line_index, 'line_geometry': line_geom})
                break  # Stop searching for lines once a match is found for the point
    # Append pts with the matched lines information
    matched_lines_df = pd.DataFrame(matched_lines)
    pts = pts.merge(matched_lines_df, left_index=True, right_on='pt_id', how='left')
    return pts
```

```python
# Input orth distace based on pts.crs units of measurement
def turner(gp_pts, gp_bds, **kwargs):
    #Default parameters
    orth_distance = kwargs.get('orth_distance', 15)
    reduced = kwargs.get('reduced', True)
    
    
    #Only keeping Points and lineStrings
    pts = gp_pts[gp_pts['geometry'].geom_type == 'Point']
    bds = gp_bds[gp_bds['geometry'].geom_type == 'LineString']
    orig_crs = gp_pts.crs
    # Finding nearest linestring and distance
    gpd.options.display_precision = 10
    pts = pts.to_crs(4326)
    bds = bds.to_crs(4326)
    df_n = gpd.sjoin_nearest(pts, bds).merge(bds, left_on="index_right", right_index=True)
    df_n["distance"] = df_n.apply(lambda r: r["geometry_x"].distance(r["geometry_y"]), axis=1)
    
    # Shortest line between point and linestring 
    p = gpd.GeoSeries(df_n["geometry_x"])
    l = gpd.GeoSeries(df_n["geometry_y"])
    shortest_line = p.shortest_line(l, align=True)
    df_n["shortest_line"] = shortest_line
     
    
    def orthogonal_points(line, distance_in_meters):
        dist = distance_in_meters/100000
        point = Point(line.coords[1][0], line.coords[1][1])
        # Calculate the slope of the line
        dx = line.coords[1][0] - line.coords[0][0]
        dy = line.coords[1][1] - line.coords[0][1]
        if dx == 0:  # If the line is vertical, swap dx and dy
            dx, dy = dy, dx
        if dx == 0: # Accounting for divide by 0
            return
        slope = dy / dx
        
        # Calculate the angle of the line
        angle = np.arctan(slope)
        
        # Calculate the coordinates of the orthogonal points
        x1 = point.x + dist * np.cos(angle + np.pi / 2)
        y1 = point.y + dist * np.sin(angle + np.pi / 2)
        x2 = point.x + dist * np.cos(angle - np.pi / 2)
        y2 = point.y + dist * np.sin(angle - np.pi / 2)
        
        return Point(x1, y1), Point(x2, y2)
    
    df_n['orthogonal_points'] = df_n.apply(lambda row: orthogonal_points(LineString(row['shortest_line']), orth_distance), axis=1)
    
    df_n[['orthogonal_point1', 'orthogonal_point2']] = df_n['orthogonal_points'].apply(pd.Series)
    
    def both_points_inside(geo, op1, op2):
        if geo.geom_type != 'LineString':
            return False
        if geo.distance(op1) < 0.000001 and geo.distance(op2) < 0.000001:
            return True
        else:
            return False

        
    
    df_n['turner_pass'] = df_n.apply(lambda row: both_points_inside(LineString(row['geometry_y']),Point(row['orthogonal_point1']),Point(row['orthogonal_point2'])), axis=1)
    
    if reduced == True:
        df_n = df_n.drop(df_n.loc[:, 'geometry_y':'orthogonal_point2'].columns, axis=1)
        
    # df_n.rename(columns={'geometry_x': 'geometry'})
    gdf = gpd.GeoDataFrame(df_n, geometry= 'geometry_x', crs="EPSG:4326")
    gdf = gdf.to_crs(orig_crs)    
    return gdf
```

```python
def drop_tiny_lines(bds, method = 'percentile', **kwargs ):
    reduced = kwargs.get('reduced', True)
    orig_crs = bds.crs
    bds = bds[bds['geometry'].geom_type == 'LineString']
    df = bds.to_crs(4326)
    df['length'] = df.length
    
    if method == 'percentile':
        percentile = kwargs.get('percentile', 0.01)
        cut_off = df.length.quantile(percentile)
        df = df[df['length'] >= cut_off]
    
    if method == 'number_of_std':
        num_dev = kwargs.get('num_dev', 2)
        cut_off = df.length.mean() - num_dev * df.length.std()
        df = df[df['length'] >= cut_off]    
    
    if method == 'length':
        meters = kwargs.get('meters', 500)
        cut_off = meters/100000
        df = df[df['length'] >= cut_off]

    if reduced == True:
        df = df.drop(df.loc[:,'length'.columns], axis=1)
        
    df_n = df.to_crs(orig_crs)
    return df_n
    
```

```python
import geopandas as gpd
from shapely.geometry import LineString, Point, Polygon
import numpy as np
import pandas as pd


# Assign characteristics to points
## first input is the shape file containing the points on the map
## second input is the shape file containing the polygons for the characteristics
## e,g assign_characteristics(addresses, school_districs)
def points_in_polygon(pts, cha, cha_name):
    # Perform spatial join with op='within'
    joined = gpd.sjoin(pts, cha, how="left", lsuffix='', rsuffix=cha_name, predicate="within")
    return joined



# Turner algorithm to match points to boundries
## Boundries must be linestrings
##

# Input orth distace based on pts.crs units of measurement
def turner(gp_pts, gp_bds, **kwargs):
	# Default parameters
	orth_distance = kwargs.get('orth_distance', 15)
	reduced = kwargs.get('reduced', True)
	unit_crs = kwargs.get('unit_crs', 3857)

	# Only keeping Points and lineStrings
	pts = gp_pts[gp_pts['geometry'].geom_type == 'Point']
	bds = gp_bds[gp_bds['geometry'].geom_type == 'LineString']
	orig_crs = gp_pts.crs
	# Finding nearest linestring and distance
	gpd.options.display_precision = 10
	pts = pts.to_crs(unit_crs)
	bds = bds.to_crs(unit_crs)
	df_n = gpd.sjoin_nearest(pts, bds).merge(bds, left_on="index_right", right_index=True)
	df_n["distance"] = df_n.apply(lambda r: r["geometry_x"].distance(r["geometry_y"]), axis=1)

	# Shortest line between point and linestring
	p = gpd.GeoSeries(df_n["geometry_x"])
	l = gpd.GeoSeries(df_n["geometry_y"])
	shortest_line = p.shortest_line(l, align=True)
	df_n["shortest_line"] = shortest_line

	def orthogonal_points(line, distance_in_meters):
		dist = distance_in_meters
		point = Point(line.coords[1][0], line.coords[1][1])
		# Calculate the slope of the line
		dx = line.coords[1][0] - line.coords[0][0]
		dy = line.coords[1][1] - line.coords[0][1]
		if dx == 0:  # If the line is vertical, swap dx and dy
			dx, dy = dy, dx
		if dx == 0:  # Accounting for divide by 0
			return
		slope = dy / dx

		# Calculate the angle of the line
		angle = np.arctan(slope)

		# Calculate the coordinates of the orthogonal points
		x1 = point.x + dist * np.cos(angle + np.pi / 2)
		y1 = point.y + dist * np.sin(angle + np.pi / 2)
		x2 = point.x + dist * np.cos(angle - np.pi / 2)
		y2 = point.y + dist * np.sin(angle - np.pi / 2)

		return Point(x1, y1), Point(x2, y2)

	df_n['orthogonal_points'] = df_n.apply(
		lambda row: orthogonal_points(LineString(row['shortest_line']), orth_distance), axis=1)

	df_n[['orthogonal_point1', 'orthogonal_point2']] = df_n['orthogonal_points'].apply(pd.Series)

	def both_points_inside(geo, op1, op2):
		if geo.geom_type != 'LineString':
			return False
		if geo.distance(op1) < 0.000001 and geo.distance(op2) < 0.000001:
			return True
		else:
			return False

	df_n['turner_pass'] = df_n.apply(
		lambda row: both_points_inside(LineString(row['geometry_y']), Point(row['orthogonal_point1']),
									   Point(row['orthogonal_point2'])), axis=1)

	if reduced == True:
		df_n = df_n.drop(df_n.loc[:, 'geometry_y':'orthogonal_point2'].columns, axis=1)

	# df_n.rename(columns={'geometry_x': 'geometry'})
	gdf = gpd.GeoDataFrame(df_n, geometry='geometry_x', crs="EPSG:4326")
	gdf = gdf.to_crs(orig_crs)
	return gdf


def drop_tiny_lines(bds, method='percentile', **kwargs):
	reduced = kwargs.get('reduced', True)
	unit_crs = kwargs.get('unit_crs', 3857)
	orig_crs = bds.crs
	bds = bds[bds['geometry'].geom_type == 'LineString']
	df = bds.to_crs(unit_crs)
	df['length'] = df.length

	if method == 'percentile':
		percentile = kwargs.get('percentile', 0.01)
		cut_off = df.length.quantile(percentile)
		df = df[df['length'] >= cut_off]

	if method == 'number_of_std':
		num_dev = kwargs.get('num_dev', 2)
		cut_off = df.length.mean() - num_dev * df.length.std()
		df = df[df['length'] >= cut_off]

	if method == 'length':
		meters = kwargs.get('meters', 500)
		cut_off = meters
		df = df[df['length'] >= cut_off]

	if reduced == True:
		df = df#.drop(df.loc[:, 'length'], axis=1)

	df_n = df.to_crs(orig_crs)
	return df_n


bds = gpd.read_file(r"C:\Users\shahi\OneDrive - University of Warwick\Plot Sizes\test_data\Boundries.shp")
pts = gpd.read_file(r"C:\Users\shahi\OneDrive - University of Warwick\Plot Sizes\test_data\Points.shp")

turned = turner(pts, bds, reduced=False)
turned
```

```python
def smooth_geometry(geom, window_size=3):
    if geom.geom_type == 'Polygon':
        exterior_coords = np.array(geom.exterior.coords)
        smoothed_exterior = moving_average(exterior_coords, window_size)
        return Polygon(smoothed_exterior)
    elif geom.geom_type == 'LineString':
        coords = np.array(geom.coords)
        smoothed_coords = moving_average(coords, window_size)
        return LineString(smoothed_coords)
    else:
        return geom

def moving_average(coords, window_size):
    coords = np.pad(coords, ((window_size//2, window_size//2), (0, 0)), mode='edge')
    smoothed = np.convolve(coords[:, 0], np.ones(window_size)/window_size, mode='valid')
    smoothed_y = np.convolve(coords[:, 1], np.ones(window_size)/window_size, mode='valid')
    return np.vstack((smoothed, smoothed_y)).T


# Apply the smoothing function
smoothed_area = area.copy()
smoothed_area['geometry'] = smoothed_area['geometry'].apply(lambda geom: smooth_geometry(geom, window_size=3))
smoothed_area.plot()
plt.show()
```

```python
# Compute the centroids of the polygons
centroids = area.geometry.centroid

# Create a new GeoDataFrame with the centroids and ids
center = gpd.GeoDataFrame(area[['id']].copy(), geometry=centroids)

# Ensure the CRS of the new GeoDataFrame is the same as the original
center.set_crs(area.crs, inplace=True)
center
```

```python

# Combine all polygons into a single MultiPolygon using unary_union
multi_poly = area.geometry.unary_union

# Create the minimum rotated bounding box
rotated_bounding_box = multi_poly.minimum_rotated_rectangle

# Create a new GeoDataFrame containing the rotated bounding box
rotated_envelope = gpd.GeoDataFrame([{'geometry': rotated_bounding_box}], crs=area.crs)

# Plotting
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the area polygons
area.plot(ax=ax, color='blue', edgecolor='black', alpha=0.5)

# Plot the rotated bounding box
rotated_envelope.plot(ax=ax, facecolor='none', edgecolor='red', linewidth=2)

# Add title and labels
ax.set_title('Polygons and Minimum Rotated Bounding Envelope')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()
```
